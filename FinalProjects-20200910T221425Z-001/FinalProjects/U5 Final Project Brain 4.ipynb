```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
df = pd.read_csv("/content/drive/My Drive/BridgeUp/Paleontology/Brain four Final project/bird_dino_data.csv")
```

```
df["Body Mass (g)"]= df["Body Mass (kg)"]*1000
```

```
df.head()
```

Brain to Body Ratio


```
dfBirdBrain = df["Brain Volume (cm3)"].loc[df["Bird vs Dino"]== "B"]
dfBirdBody = df["Body Mass (g)"].loc[df["Bird vs Dino"]== "B"]
BirdRatio= dfBirdBrain/dfBirdBody
Birdbtbavg = BirdRatio.mean()
print(Birdbtbavg)
```

```
dfDinoBrain = df["Brain Volume (cm3)"].loc[df["Bird vs Dino"]== "D"]
dfDinoBody = df["Body Mass (g)"].loc[df["Bird vs Dino"]== "D"]
DinoRatio= dfDinoBrain/dfDinoBody
Dinobtbavg = DinoRatio.mean()
print(Dinobtbavg)
```

Cerebrum to Whole brain

```
dfBirdCerebrum = df["Cerebrum (cm3)"].loc[df["Bird vs Dino"]== "B"]
dfBirdWhole = df["Brain Volume (cm3)"].loc[df["Bird vs Dino"]== "B"]
Birdctw = dfBirdCerebrum/dfBirdWhole
birdctwavg = Birdctw.mean()
print(birdctwavg)
```

```
dfDinoCerebrum = df["Cerebrum (cm3)"].loc[df["Bird vs Dino"]== "D"]
dfDinoWhole = df["Brain Volume (cm3)"].loc[df["Bird vs Dino"]== "D"]
Dinoctw = dfDinoCerebrum/dfDinoWhole
Dinoctwavg = Dinoctw.mean()
print(Dinoctwavg)
```

Bar graph

```
birdbtbpercent = Birdbtbavg*100
Dinobtbpercent = Dinobtbavg*100
birdctwpercent = birdctwavg*100
Dinoctwpercent = Dinoctwavg*100
print(birdbtbpercent)
print(Dinobtbpercent)
print(birdctwpercent)
print(Dinoctwpercent)
```

Bird to Brain Graph

```
plt.bar([1,2], [birdbtbpercent, Dinobtbpercent], color = ["paleturquoise", "darkcyan"])
plt.xlabel("Animals")
plt.xticks([1,2], ["Bird Ratio", "Dino Ratio"])
plt.ylabel("Percentage (%)")
plt.title ("Brain to Body Mass Ratio")
plt.scatter([], [], color = "paleturquoise", label= "66.49%")
plt.scatter([], [], color = "darkcyan", label= "46.95%")
plt.legend()
plt.show()
```

```
plt.bar([1,2], [birdctwpercent, Dinoctwpercent], color = ["thistle", "darkmagenta"])
plt.xlabel("Animals")
plt.xticks([1,2], ["Bird Ratio", "Dino Ratio"])
plt.ylabel("Percentage (%)")
plt.title ("Cerebrum to Whole Brain Ratio")
plt.scatter([], [], color = "thistle", label= "66.49%")
plt.scatter([], [], color = "darkmagenta", label= "46.95%")
plt.legend()
plt.show()
```
